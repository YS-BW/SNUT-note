###  布尔索引：过滤数据元素


```python
mb=np.array([[1,2],[3,4],[5,6]])
print(mb)
```

```python
bool_idx=mb>2  #拿到一个布尔数组，其中元素True或False代表相应位置的元素是否大于2
print(bool_idx)
mm=mb[bool_idx]  #索引
print(mm,mm.shape)
```

#### 注意：
- 布尔数组必须与要索引的数组形状相同
- 布尔数组返回的新数组是原数组的副本，与原数组不共享数据空间，新数组的修改不会影响原数组，这就是所谓的深层复制。

## 六、Numpy 数组：合并、分割、广播 

### stacking array 合并

```python
a1=np.full((1,5),1.0)
b1=np.full((1,5),2.0)
svab=np.vstack((a1,b1)) # 垂直合并（行方向）
print(svab)
print(svab.shape)
```

```python
a2=np.full((3,2),1.0)
print(a2)
b2=np.full((3,2),2.0)
print(b2)
shab=np.hstack((a2,b2))  #水平合并（列方向）
print(shab)
print(shab.shape)
```

###   concatenate  显性定义合并的方向

```python

cvab=np.concatenate((a2,b2),axis=0) #和vstack相同
display(cvab,cvab.shape)

chab=np.concatenate((a2,b2),axis=1) #和hstack相同
display(chab,chab.shape)
```

### 分割数组  split

```python
big_ar=np.arange(36).reshape(9,4)
print(big_ar)
```

```python
m1,m2,m3=np.vsplit(big_ar,3) #将一个数组垂直分割成多个子数组（按行）
print(m1)
print("*****************")
print(m2)
print("*****************")
print(m3)

```

```python
m4,m5=np.hsplit(big_ar,2) #将一个数组水平分割成多个子数组（按列）
print(m4)
print("************")
print(m5)
```

###  Tile  重复

```python
m1=np.arange(4)
print(m1)
print("******************")

m2=np.tile(m1,(4,1))  #重复4个m1数组，并将它们合并
print(m2)

print("******************")

m3=np.tile(m1,(4,2))
print(m3)
```

###  广播 Broadcasting

- 数组无需循环遍历就可以对每个元素进行批量的算术运算。
- 形状相同时，是位置相同的元素进行运算；
- 形状不同时，Numpy设计了一种广播机制，较小的数组，在横向或纵向上进行一定次数的重复，使其与较大的数组拥有相同的维度

### 数组的文件输入与输出

- 数组保存为文本文件、csv文件，并读取，numpy.savetxt()、numpy.loadtxt()
- 数组保存为二进制文件，并读取,numpy.save(),numpy.load()


```python
data=np.arange(10).reshape(2,5)
print(data)
np.savetxt('txtout.txt',data,fmt='%d')
data = np.random.randint(10,size=(2,5))
print(data)
np.savetxt('csvoout.csv',data,fmt='%d',delimiter=',')
```

```python
import numpy as np
student_id = "202318084032"
txt_file = f"{student_id}.txt"
np.savetxt(txt_file, data, fmt='%d')
csv_file = f"{student_id}.csv"
np.savetxt(csv_file, data, fmt='%d', delimiter=',')
loaded_txt = np.loadtxt(txt_file, dtype=int)
loaded_csv = np.loadtxt(csv_file, dtype=int, delimiter=',')
print(f"原始数据:\n{data}")
print(f"\n从 {txt_file} 读取的数据:\n{loaded_txt}")
print(f"\n从 {csv_file} 读取的数据:\n{loaded_csv}")
```
