# 2️⃣多图布局

**均匀分布**
> subplot()

```python
# 画布
fig = plt.figure(figsize=(8,5))
x = np.linspace(-np.pi,np.pi,30)
y = np.sin(x)
# 子图1
axes1 = plt.subplot(221)# 二行二列第一个
axes1.plot(x,y)
axes1.set_title('子图1')
# 子图2
axes2 = plt.subplot(222)# 2⾏2列的第2个⼦视图
axes2.plot(x,y)
axes2.set_title('子图2')
# 子图3
axes3 = plt.subplot(223)# 2⾏2列的第3个⼦视图
axes3.plot(x,y)
axes3.set_title('子图3')
# 子图2
axes4 = plt.subplot(224)# 2⾏2列的第4个⼦视图
axes4.plot(x,y)
axes4.set_title('子图4')
# 自动调整布局
fig.tight_layout()
```

**不均匀分布**
> subplo()

```python
plt.figure(figsize=(8,5))

x = np.linspace(-np.pi,np.pi,30)
y = np.sin(x)

axes1 = plt.subplot(2,2,1)
axes1.plot(x,y,color = 'red')

axes2 = plt.subplot(2,2,2)
lines = axes2.plot(x,y)
lines[0].set_marker('*') # lines[0]是第一个,也是唯一一个

axes3 = plt.subplot(2,1,2)
axes3.plot(x,np.sin(x*x))
```

**图形嵌套**
> add_supplot()

```python
fig = plt.figure(figsize=(8,5))

axes1 = fig.add_subplot(1,1,1) # 和subplot相比,不自动创建figure对象
axes1.plot([1,3]) #默认为y点坐标(1,3)

axes2 = fig.add_subplot(2,2,1,facecolor = 'pink')
axes2.plot([1,3])
```

> 使用axes()和add_axes()

```python
fig = plt.figure(figsize=(8,5))

x = np.linspace(0,2*np.pi,30)
y = np.sin(x)
plt.plot(x,y)

axes1 = plt.axes([0.55,0.55,0.3,0.3])# 新坐标轴的位置和大小
axes1.plot(x,y,color = 'g')

axes2 = fig.add_axes([0.2,0.2,0.25,0.25])# 新坐标轴的位置和大小
axes2.plot(x,y,color = 'r')
```

**均匀布局**
> 使用subplots()

```python
x = np.linspace(0,2*np.pi)

#创建一个 3 行 3 列的子图网格（共 9 个子图），
#并返回 Figure 对象（fig）和 子图对象数组（ax）
fig,ax = plt.subplots(3,3) 
ax1,ax2,ax3 = ax # 三行
ax11 , ax12 , ax13 = ax1 # 第一,二,三行⬇️
ax21 , ax22 , ax23 = ax2
ax31 , ax32 , ax33 = ax3

#设置画布
fig.set_figheight(5)
fig.set_figwidth(8)

#draw
ax11.plot(x,np.sin(x))
ax12.plot(x,np.cos(x))
ax13.plot(x, np.tanh(x))
ax21.plot(x, np.tan(x))
ax22.plot(x, np.cosh(x))
ax23.plot(x, np.sinh(x))
ax31.plot(x, np.sin(x) + np.cos(x))
ax32.plot(x, np.sin(x*x) + np.cos(x*x))
ax33.plot(x, np.sin(x)*np.cos(x))

#适应布局
plt.tight_layout()
#渲染
plt.show()
```

**双轴显示**
> 

```python
plt.figure(figsize=(8,5))

x = np.linspace(0,10,100)

axes1 = plt.gca() #获得轴域
axes1.plot(x,np.exp(x),color = 'red')#draw
axes1.set_xlabel('time')#x轴
axes1.set_ylabel('exp',color='red')#y轴
axes1.tick_params(axis = 'y',labelcolor = 'blue')

axes2 = axes1.twinx()#共用x轴
axes2.set_ylabel('sin',color = 'blue')
axes2.plot(x,np.sin(x),color = 'blue')
axes2.tick_params(axis = 'y',labelcolor = 'blue')

plt.tight_layout()
plt.show()
```

# 3️⃣Matplotlib绘图属性设置

**常用函数**
| Pyplot函数 | API方法 | 描述 |
| --- | --- | --- |
| text() | mpl.axes.Axes.text() | 在Axes对象的任意位置添加文字 |
| xlabel() | mpl.axes.Axes.set_xlabel() | 为X轴添加标签 |
| ylabel() | mpl.axes.Axes.set_ylabel() | 为Y轴添加标签 |
| title() | mpl.axes.Axes.set_title() | 为Axes对象添加标题 |
| legend() | mpl.axes.Axes.legend() | 为Axes对象添加图例 |
| annotate() | mpl.axes.Axes.annotate() | 为Axes对象添加注释（箭头可选） |
| suptitle() | mpl.figure.Figure.suptitle() | 为Figure对象添加中心化的标题 |


```python
fig = plt.figure(figsize=(8,5))

x = np.linspace(0,2*np.pi)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))

# 图例
plt.legend(['sin','cos'],
            fontsize = 18,
            loc = 'center',# 居中
            ncol = 2,# 列数
            bbox_to_anchor=[0,1.05,1,0.2]
            )
plt.show()
```

```python
fig = plt.figure(figsize=(8,5))

x = np.linspace(0,2*np.pi,20)
y1 = np.sin(x)
y2 = np.cos(x)

# c : color 线颜⾊
# marker: 标记的样式或点的样式
# mfc: marker face color 标记的背景颜⾊
# ls : line style 线的样式
# lw: line width 线的宽度
# label: 线标签（图例中显示）

plt.plot(x,y1,c = 'r',marker = 'o',ls = '--',lw = 1,label = 'sinx',mfc = 'y')
plt.plot(x,y2,c = 'b',marker = '*',ls = '-',lw = 2,label = 'cosx',mfc = 'g')
plt.plot(x, y1-y2, c='y', marker='^', ls='-', lw=3, label='sinx-cosx',mfc='b', alpha=0.5)
plt.plot(x, y1+y2, c='orange', marker='>', ls='-.', lw=4, label='sinx+cosx',
     mfc='y',
     markersize=10, # 点⼤⼩
     markeredgecolor='green', #点边缘颜⾊
     markeredgewidth=2 # 点边缘宽度
     )
plt.legend()
```

```python
# **Line Styles**
# ============= ===============================
# character description
# ============= ===============================
# ``'-'`` solid line style
# ``'--'`` dashed line style
# ``'-.'`` dash-dot line style
# ``':'`` dotted line style
# ============= ===============================
# **Colors**
# The supported color abbreviations are the single letter codes
# ============= ===============================
# character color
# ============= ===============================
# ``'b'`` blue
# ``'g'`` green
# ``'r'`` red
# ``'c'`` cyan
# ``'m'`` magenta
# ``'y'`` yellow
# ``'k'`` black
# ``'w'`` white
# ============= ===============================
# **Markers**
# ============= ===============================
# character description
# ============= ===============================
# ``'.'`` point marker
# ``','`` pixel marker
# ``'o'`` circle marker
# ``'v'`` triangle_down marker
# ``'^'`` triangle_up marker
# ``'<'`` triangle_left marker
# ``'>'`` triangle_right marker
# ``'1'`` tri_down marker
# ``'2'`` tri_up marker
# ``'3'`` tri_left marker
# ``'4'`` tri_right marker
# ``'8'`` octagon marker
# ``'s'`` square marker
# ``'p'`` pentagon marker
# ``'P'`` plus (filled) marker
# ``'*'`` star marker
# ``'h'`` hexagon1 marker
# ``'H'`` hexagon2 marker
# ``'+'`` plus marker
# ``'x'`` x marker
# ``'X'`` x (filled) marker
# ``'D'`` diamond marker
# ``'d'`` thin_diamond marker
# ``'|'`` vline marker
# ``'_'`` hline marker
# ============= ===============================
```

**坐标轴刻度**
> xticks
> yticks

```python
x = np.linspace(0,10)
y = np.sin(x)
plt.plot(x,y)

#设置x轴y轴刻度
plt.xticks(np.arange(0,11,1))
plt.yticks([-1,0,1])
plt.show()
```

```python
x = np.linspace(0,10)
y = np.sin(x)
plt.plot(x,y)

plt.yticks(ticks=[-1,0,1],
           labels = ['min','0','max'], # 刻度对应的标签
          fontsize = 20,
            ha = 'right',#水平对齐方式
            color = 'blue')
plt.xticks(ticks=np.arange(0,11,1),fontsize=20,color = 'red')
plt.show()
```

**坐标轴范围**
> xlim
> ylim

```python
x = np.linspace(0,2*np.pi)
y = np.sin(x)
plt.plot(x,y,c = 'r')

#设置x轴范围
plt.xlim(-2,8)
#y
plt.ylim(-2,2)

plt.show()
```

**坐标轴配置**
> axis

```python
x = np.linspace(0,2*np.pi)
y = np.sin(x)
plt.plot(x,y,color = 'r')

# 坐标轴范围：[xmin, xmax, ymin, ymax]
plt.axis([-2,8,-2,2])

plt.axis('square')
# 选项
# off ： 不显示坐标轴
# equal: 让x轴和y轴 刻度距离相等
# scaled：⾃动缩放坐标轴和图⽚适配
# tight：紧凑型⾃动适配图⽚
# square：x轴和y轴宽⾼相同
plt.show()
```

**标题和网络**
> title
> grid

```python
x = np.linspace(0,10)
y = np.sin(x)
plt.plot(x,y)

plt.title('sin',fontsize= 20,loc='center')
plt.suptitle('fsin',y = 1.1,fontsize = 30)

plt.grid(ls='--',lw = 0.5,c = 'gray',axis='y')#添加坐标轴网格线
plt.show()
```

**标签**
> xlabel
> ylabel

```python
x = np.linspace(0,10)
y = np.sin(x)
plt.plot(x,y)

# 坐标轴标签
plt.xlabel('y = sin(X)',
              fontsize = 20,
              rotation = 0
          )
plt.ylabel('y = sin(X)',
              rotation = 90,
              horizontalalignment = 'right',# ⽔平对⻬⽅式
              fontsize = 20)
plt.title('正弦曲线')
```

**文本**
> text

```python
plt.figure(figsize=(8,5))
x= np.linspace(0,10,10)
y = np.array([60,30, 20, 90, 40, 60, 50, 80, 70, 30])
plt.plot(x,y,ls = '--',marker= 'o')

for a,b in zip(x,y):
    plt.text(
            x = a+0.3,#表示向右偏移 0.3 个单位
            y = b+0.5,#表示向上偏移 0.5 个单位
            s=b,# ⽂字内容
            ha = 'center',#水平对齐
            va = 'center',#垂直对齐
            fontsize = 14,
            color = 'r'
    )
plt.show()
```

**注释**
> annotate

```python
plt.figure(figsize=(8,5))

x = np.linspace(0,10,10)
y = np.array([60, 30, 20, 90, 40, 60, 50, 80, 70, 30])
plt.plot(x,y,ls = '--',marker = 'o')

plt.annotate(
                text='max',
                xy = (3,90),
                xytext=(1,80),
                arrowprops={
                         'width': 2, # 箭头线的宽度
                         'headwidth': 8, # 箭头头部的宽度
                         'facecolor': 'blue' # 箭头的背景颜⾊
                },
)
```

**保存图片**
> savefig

```python
f = plt.figure(figsize=(8,5))

x = np.linspace(0,2*np.pi)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))

f.savefig(
            fname='pic_name2.png', # ⽂件名：png、jpg、pdf
             dpi=100, # 保存图⽚像素密度
             facecolor='pink', # 背景颜⾊
             edgecolor='lightgreen', # 边界颜⾊
             bbox_inches='tight', # 保存图⽚完整
             pad_inches=1 # 内边距
        
)
```
